```python
df_dummies=pd.get_dummies(df['Ethnicity']) #one hot encoding the ethnicity variables
df_dummies.columns=['AI', 'AN', 'BK', 'HC', 'TE', 'WE'] #renaming them to match the data dictionary
df_onehot=df.join(df_dummies) #joining the one hot dataframe with the other features
```

```python
df_onehot
```

```python
from sklearn.model_selection import cross_val_score
from numpy import absolute
```

```python
X=df_onehot.drop(columns=['TOSLSTOT', 'TOSLS1', 'TOSLS2', 'TOSLS3', 'TOSLS4', 'TOSLS5', 'TOSLS6', 'TOSLS7', 'TOSLS8', 'TOSLS9', 'BRAINS1','BRAINS2','BRAINS3','BRAINS4','BRAINS5', 'Gender', 'Teacher', 'Period', 'Student ID', 'SORT', 'Ethnicity', 'TOT']) #we might try one-hot encoding some categoricals
y=df['TOSLSTOT']
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
regressor = xgb.XGBRegressor(n_estimators=100,reg_lambda=1,gamma=0,max_depth=3,objective ='reg:squarederror')
RMSE_list=[]
kf = KFold(n_splits=10, random_state=42, shuffle=True)
RMSE_list = cross_val_score(regressor, X, y, scoring='neg_mean_absolute_error',cv=kf, n_jobs=-1)  
print("The RMSE is " + str(mean(absolute(RMSE_list))))
```

```python
#this is the model from the paper from the multilinear regression done by Chandler et al. quoted
X_test_paper=df_onehot.drop(columns=['CTE', 'TOSLSTOT', 'TOSLS1', 'TOSLS2', 'TOSLS3', 'TOSLS4', 'TOSLS5', 'TOSLS6', 'TOSLS7', 'TOSLS8', 'TOSLS9', 'BRAINS1','BRAINS2','BRAINS3','BRAINS4','BRAINS5', 'Gender', 'Teacher', 'Period', 'Student ID', 'SORT', 'Ethnicity', 'TOT']) 
y_test=df_onehot['TOSLSTOT']
chandler_weights=[.807,-1.049,.936,1.772,.087,.119,.020,.668,.041,2.511,1.231,.858,-0.893,1.065,0.234]
y_test_paper_preds=X_test_paper @ chandler_weights + -14.212
RMSE = np.sqrt(mean_squared_error(y_test, y_test_paper_preds))
print("The RMSE is " + str(RMSE))
```
